This python jupyter notebook utilizes python libraries such as matplotlib, seaborn, and padas to perform exploratory data anlysis on a set of Healthcare Inurance data and goes with SQL queries and outputs that I did as part of this EDA project.

Importing the necessary libraries/modules for data analysis and loading the dataset, then printing out the first couple of rows of data.

```python
#importing the necessary libraries for data analysis and visualization and setting the style for the plots we use matplotlib (plt) , and seaborn(sns)
#when to use plt v. sns: plt is used for basic plotting and sns is used for more advanced statistical visualizations
import pandas as pd #pandas for data manipulation and analysis
import matplotlib.pyplot as plt #matplotlib is python's library for plotting graphs and colorful visualizations
import seaborn as sns #seaborn is a library based on matplotlib that provides a high-level interface for drawing attractive statistical graphics

# Set Seaborn style
sns.set(style='whitegrid') #sns.set() is used to set the aesthetic style of the plots so u always need this function before plotting stuff

# Load the dataset
df = pd.read_csv('healthcare_insurance_dataset_csv.csv')  # pandas read_csv() function is used to read a comma-separated values(csv) file into a DataFrame, file needs to be in same folder as ipynb notebook

#Display the first few rows of the dataset using the head() function of the DataFrame which is df
df.head() #df is the DataFrame containing the dataset, and head() returns the first five rows of the dataset by default
```

This bar plot visualizes the overall distrbution of insurance charges to get an overall sense of the data and get a better sense of trends, patterns, outliers, shape, and skewness. This lets us see the most frequent and recurring insruance costs in the US as well as the least.

The data shows that most insurance charges are concentrated under $15,000, with a few extreme outliers exceeding $60,000. This right-skewed distribution highlights how a small group of high-cost cases drives overall expenses, an insight useful for insurers in risk management and policy decisions.

```python
#Visualizing the overall distribution of insurance charges to get a overall sense of the data and maybe see outliers, shape, skewnes, etc like other stats vars. basically
#plt is connected to the matplotlib library, which is used for plotting graphs , so we use plt. a function 
plt.figure(figsize=(10, 6),facecolor="#e7d4dc")  #figsize is used to set the size of the figure in inches, in this case, 10 inches wide and 6 inches tall
sns.histplot(df['charges'], kde=True, color='hotpink') #kde means Kernel Density Estimate, which is a way to estimate the probability density function of a continuous random variable, color is used to set the color of the histogram bars
plt.title('Distribution of Insurance Charges') #title the plot 
plt.xlabel('Charges') #xlabel is used to set the label for the x-axis
plt.ylabel('Frequency') #ylabel is used to set the label for the y-axis
plt.show() #print it out 
```

This scatterplot compares the distribution of insurance charges for smokers and non smokers. We can see that overall, smokers have a higher healthcare insurance charges in the United States.

The chart shows that smokers consistently face much higher insurance charges than non smokers across all ages. While charges rise with age for everyone, the gap between smokers and non smokers widens significantly, highlighting smoking as a major driver of healthcare costs.

```python
#Comparing the distrbution of insruance charges by smokers versus non smokers and we should be able to see that smokers have higher insurance charges on average 
plt.figure(figsize=(10, 6),facecolor = "#430A61") #plt.figure() is used to create a new figure for plotting, figsize is used to set the size of the figure in inches, in this case, 10 inches wide and 6 inches tall
sns.scatterplot(x='age', y='charges', hue='smoker', data=df, palette='pink') #sns.scatterplot() is used to create a scatter plot, x and y are the variables to be plotted on the x-axis and y-axis respectively, hue is used to color the points based on the 'smoker' column, data is the DataFrame containing the data, palette is used to set the color palette for the plot
plt.title('Insurance Charges by Age and Smoking Status', color = "white") #title the plot, color is used to set the color of the title

plt.tick_params(axis='x', colors='white')  # Change x-axis tick label color to white
plt.tick_params(axis='y', colors='white')  # Change y-axis tick label color to white, for tick labels you have to use colors not color because it is a list of colors for each tick label


plt.xlabel('Age',color = "white") #xlabel is used to set the label for the x-axis, color is used to set the color of the label
plt.ylabel('Charges',color = "white") #ylabel is used to set the label for the y-axis, color is used to set the color of the label
plt.legend(title='Smoker') #legend is used to add a legend to the plot, title is used to set the title of the legend, color is used to set the color of the legend text


plt.show()
```

This is a barplot that shows the average insurance charges by region.

The Southeast region shows the highest average insurance charges, while the Southwest and Northwest are lower. This regional variation highlights how geographic factors can influence healthcare costs and may guide pricing or policy decisions.

```python
import numpy as np  # Import NumPy for statistical functions like mean

plt.figure(figsize=(8, 5))  # Set the size of the figure (width=8, height=5)

sns.barplot(
    x='region',                 # x-axis will represent the region column
    y='charges',                # y-axis will represent the charges column
    hue='region',               # Use different colors for each region (same as x to enable pastel palette)
    data=df,                    # Data source is the dataframe named df
    estimator=np.mean,          # Use NumPy's mean function to calculate average charges
    palette='pastel',           # Apply a pastel color palette for aesthetics
    legend=False                # Disable the legend since hue and x are the same (avoids redundancy)
)

plt.title('Average Insurance Charges by Region')  # Set the title of the chart
plt.ylabel('Average Charges')                     # Label the y-axis
plt.xlabel('Region')                              # Label the x-axis
plt.show()                                        # Display the chart




```

This is a boxplot which compares key statistics for insurance charges for smokers versus non smokers. Here we can clearly see the differences in the highest/lowest values for both categories as well as the median values.


The boxplot shows that smokers consistently face higher insurance charges, with a much wider spread of costs compared to non smokers. This indicates that smoking not only increases average expenses but also contributes to more variability and extreme outliers in healthcare costs.

```python
#this code will compare the median and highest and lowest insurance charges for smokers and non-smokers using a box plot
#box plots are used to visualize the distribution of a dataset and identify outliers, they show the median, quartiles, and potential outliers in the data
plt.figure(figsize=(8, 10))
sns.boxplot(x='smoker',y='charges',data=df,palette=['green', 'red']  # First category green, second red for smokers and non-smokers
)
plt.title('Insurance Charges by Smokers vs Non-Smokers')
plt.xlabel('Smoker Status ')
plt.ylabel('Charges')
plt.show()
```

This boxplot shows the distribution of insurance charges based on the number of children in a family.

The plot shows that insurance charges do not vary dramatically with the number of children. While families with more children (3–4) display slightly higher medians and wider spreads, the overall pattern suggests that the number of children has less impact on charges compared to other factors like smoking or age. But maybe we can make a inference that people with more children have different family coverage plans which can drive overall costs lower.

```python
# this boxplot will show the distrribution of insurance charges based on the number of children in a family
plt.figure(figsize=(12, 10))
sns.boxplot(x='children', y='charges', data=df, palette='Set2')  # Distribution of charges for each family size
plt.title('Distribution of Charges by Number of Children')
plt.xlabel('Number of Children')
plt.ylabel('Charges')
plt.show()
```

This analysis highlights that smoking status and age are the strongest drivers of higher insurance charges, while regional and family-size differences are less significant. By using these insights, insurers can better segment risk, adjust premiums, and design targeted wellness programs that reduce costly claims. Leveraging data in this way supports both healthier outcomes for customers and profitable growth within insurance products.
